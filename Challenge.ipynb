```python
# Import dependencies
import pandas as pd
import numpy as np
import requests

from citipy import citipy
from config import weather_api_key
```

```python
# Import the dataset
flights_inf = 'challenge_dataset.csv'

# Create the flight Dataframe
flights_df = pd.read_csv(flights_inf)

flights_df.tail(10)
```

```python
flights_df= flights_df.drop_duplicates(subset='flight_num')

flights_order = ["origin","destination", "airline","flight_num","origin_name",'origin_latitude','origin_longitude',"destination_name",'destination_latitude','destination_longitude']
flights_df = flights_df[flights_order]
```

```python
flights_df = flights_df.rename(columns={'origin': 'Origin', 'destination': 'Destination','airline':'Airline', 'flight_num':'Flight', 'origin_name':'Origin Airport', 'destination_name':'Destination Airport'})
flights_df.tail(10)
```

```python
row_num = flights_df.shape[0]
print("Number of rows on the DataFrame:", row_num)
```

```python
# Get unique origin and destination combinations
unique_flights = flights_df.groupby(['Origin', 'Destination']).size().reset_index(name='Conteo')

# Get the latitudes and longitudes for each unique combination
for index, row in unique_flights.iterrows():
    origin = row['Origin']
    destination = row['Destination']
    latitude_origin = flights_df.loc[flights_df['Origin'] == origin, 'origin_latitude'].iloc[0]
    longitude_origen = flights_df.loc[flights_df['Origin'] == origin, 'origin_longitude'].iloc[0]
    latitude_destination = flights_df.loc[flights_df['Destination'] == destination, 'destination_latitude'].iloc[0]
    longitude_destination = flights_df.loc[flights_df['Destination'] == destination, 'destination_longitude'].iloc[0]
    unique_flights.at[index, 'origin_latitude'] = latitude_origin
    unique_flights.at[index, 'origin_longitude'] = longitude_origen
    unique_flights.at[index, 'destination_latitude'] = latitude_destination
    unique_flights.at[index, 'destination_longitude'] = longitude_destination
    
df_unique_flights = pd.DataFrame(unique_flights)
df_unique_flights.tail(20)
```

```python
# Get each location from both the origin and destination columns
places = pd.concat([df_unique_flights['Origin'], df_unique_flights['Destination']])

# Remove duplicate values ​​from places column
unique_places = places.drop_duplicates()

# Create a list to store temporary DataFrames
dfs = []

# Iterate over unique locations to get their latitudes and longitudes
for place in unique_places:
    latitude = df_unique_flights.loc[df_unique_flights['Origin'] == place, 'origin_latitude'].iloc[0] if place in df_unique_flights['Origin'].values else df_unique_flights.loc[df_unique_flights['Destination'] == place, 'destination_latitude'].iloc[0]
    longitude = df_unique_flights.loc[df_unique_flights['Origin'] == place, 'origin_longitude'].iloc[0] if place in df_unique_flights['Origin'].values else df_unique_flights.loc[df_unique_flights['Destination'] == place, 'destination_longitude'].iloc[0]
 
    df_temp = pd.DataFrame({'place': [place], 'latitude': [latitude], 'longitude': [longitude]})
    dfs.append(df_temp)
    df_places = pd.concat(dfs, ignore_index=True)

# Print the DataFrame of places with latitudes and longitudes
print(df_places)
```

```python
# Create a list of tuples with latitudes and longitudes from df_places
lats_lngs = [(row.latitude, row.longitude) for row in df_places.itertuples(index=False)]

lats_lngs
```

```python
# Build URL
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key
```

```python
# Create an empty list to hold the weather data
city_data = []

# Print the beginning of the logging
print("Beginning Data Retrieval")
print("-----------------------------")

# Create counters
record_count = 1
set_count = 1

# Loop through all the coordinates in the list
for lat, lon in lats_lngs:
    
    # Create endpoint URL with latitude and longitude for weather data
    weather_url = f"{url}&lat={lat}&lon={lon}"

    # Log the URL, record, and set numbers and the city
    print(f"Processing Record {record_count} of Set {set_count} | Latitude: {lat}, Longitude: {lon}")
    record_count += 1

    # Run an API request for weather data for each set of coordinates
    try:
        # Parse the JSON and retrieve weather data
        weather_response = requests.get(weather_url).json()
        
        # Parse out the needed weather data
        city_max_temp = weather_response["main"]["temp_max"]
        city_humidity = weather_response["main"]["humidity"]
        city_clouds = weather_response["clouds"]["all"]
        city_wind = weather_response["wind"]["speed"]
        city_country = weather_response["sys"]["country"]
        city_description = weather_response["weather"][0]["description"]
        
        # Get city name using reverse geocoding with the same coordinates
        reverse_geocoding_url = f"http://api.openweathermap.org/geo/1.0/reverse?lat={lat}&lon={lon}&limit=1&appid={weather_api_key}"
        reverse_geocoding_response = requests.get(reverse_geocoding_url).json()
        city_name = reverse_geocoding_response[0]["name"]
        
        # Append the city information into city_data list
        city_data.append({"City": city_name,
                          "Lat": lat,
                          "Lon": lon,
                          "Max Temp": city_max_temp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_wind,
                          "Country": city_country,
                          "Current Description": city_description})
        
    except Exception as e:
        print("Error occurred:", e)
        pass

# Indicate that Data Loading is complete
print("-----------------------------")
print("Data Retrieval Complete")
print("-----------------------------")
```

```python
city_data_df = pd.DataFrame(city_data)
city_data_df
```

```python
city_data_order = ["City", "Country", "Lat", "Lon", "Max Temp", "Humidity", "Cloudiness", "Wind Speed", "Current Description"]
city_data_df = city_data_df[city_data_order]
city_data_df.head(10)
```

```python
# Create a list to store the corresponding locations
places_match = []

# Iterate over rows of city_data_df
for index, row in city_data_df.iterrows():
    latitud = row['Lat']
    longitud = row['Lon']
    # Filter df_places rows matching exact coordinates
    matching_rows = df_places[(df_places['latitude'] == latitud) & (df_places['longitude'] == longitud)]
    # If a match is found, get the corresponding location
    if not matching_rows.empty:
        place_match = matching_rows.iloc[0]['place']
    else:
        place_match = None 
    places_match.append(place_match)

# Add the list of corresponding places as a new column in city_data_df
city_data_df['IATA code'] = places_match
city_data_df
```

```python
city_data_order = ["IATA code", "City", "Max Temp", "Humidity", "Cloudiness", "Wind Speed", "Current Description"]
city_data_df = city_data_df[city_data_order]
city_data_df.head(10)
```

```python
# Merge with climatic information for origin
flights_with_origin_weather = flights_df.merge(city_data_df, left_on='Origin', right_on='IATA code', how='left', suffixes=('', '_origin'))

# Merge with climatic information for destination
flights_with_weather = flights_with_origin_weather.merge(city_data_df, left_on='Destination', right_on='IATA code', suffixes=('', '_destination'), how='left')

flights_with_weather
```

```python
flights_with_weather_order = ["Origin","Destination","Airline","Flight","Origin Airport","City","Max Temp","Humidity","Cloudiness","Wind Speed","Current Description","Destination Airport","City_destination","Max Temp_destination","Humidity_destination","Cloudiness_destination","Wind Speed_destination","Current Description_destination"]
flights_with_weather = flights_with_weather[flights_with_weather_order]
flights_with_weather.head(10)
```

```python
flights_with_weather = flights_with_weather.rename(columns={'City':'Origin city','Max Temp':'Origin temperature','Humidity':'Origin humidity','Cloudiness':'Origin cloudiness','Wind Speed':'Origin wind speed','Current Description':'Origin current description','City_destination':'Destination city', 'Max Temp_destination':'Destination temperature','Humidity_destination':'Destination humidity','Cloudiness_destination':'Destination cloudiness','Wind Speed_destination':'Destination wind speed','Current Description_destination':'Destination current description'})
flights_with_weather.tail(10)
```
