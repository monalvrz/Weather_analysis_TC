```python
# Import dependencies
import pandas as pd
import numpy as np
import requests

from citipy import citipy
```

```python
# Import the dataset
flights_inf = 'challenge_dataset.csv'

# Create the flight Dataframe
flights_df = pd.read_csv(flights_inf)

flights_df.head(10)
```

```python
# Get unique origin and destination combinations
unique_flights = flights_df.groupby(['origin_iata_code', 'destination_iata_code']).size().reset_index(name='Conteo')

# Get the latitudes and longitudes for each unique combination
for index, row in unique_flights.iterrows():
    origin = row['origin_iata_code']
    destination = row['destination_iata_code']
    latitude_origin = flights_df.loc[flights_df['origin_iata_code'] == origin, 'origin_latitude'].iloc[0]
    longitude_origen = flights_df.loc[flights_df['origin_iata_code'] == origin, 'origin_longitude'].iloc[0]
    latitude_destination = flights_df.loc[flights_df['destination_iata_code'] == destination, 'destination_latitude'].iloc[0]
    longitude_destination = flights_df.loc[flights_df['destination_iata_code'] == destination, 'destination_longitude'].iloc[0]
    unique_flights.at[index, 'origin_latitude'] = latitude_origin
    unique_flights.at[index, 'origin_longitude'] = longitude_origen
    unique_flights.at[index, 'destination_latitude'] = latitude_destination
    unique_flights.at[index, 'destination_longitude'] = longitude_destination
    
df_unique_flights = pd.DataFrame(unique_flights)
df_unique_flights.tail(20)
```

```python
# Get each location from both the origin and destination columns
places = pd.concat([df_unique_flights['origin_iata_code'], df_unique_flights['destination_iata_code']])

# Remove duplicate values ​​from places column
unique_places = places.drop_duplicates()

# Create a list to store temporary DataFrames
dfs = []

# Iterate over unique locations to get their latitudes and longitudes
for place in unique_places:
    latitude = df_unique_flights.loc[df_unique_flights['origin_iata_code'] == place, 'origin_latitude'].iloc[0] if place in df_unique_flights['origin_iata_code'].values else df_unique_flights.loc[df_unique_flights['destination_iata_code'] == place, 'destination_latitude'].iloc[0]
    longitude = df_unique_flights.loc[df_unique_flights['origin_iata_code'] == place, 'origin_longitude'].iloc[0] if place in df_unique_flights['origin_iata_code'].values else df_unique_flights.loc[df_unique_flights['destination_iata_code'] == place, 'destination_longitude'].iloc[0]
 
    df_temp = pd.DataFrame({'place': [place], 'latitude': [latitude], 'longitude': [longitude]})
    dfs.append(df_temp)
    df_places = pd.concat(dfs, ignore_index=True)

# Print the DataFrame of places with latitudes and longitudes
print(df_places)
```

```python
# Create a list of tuples with latitudes and longitudes from df_places
lats_lngs = [(row.latitude, row.longitude) for row in df_places.itertuples(index=False)]

lats_lngs
```

```python
# Create a list for holding the cities.
cities = []
# Identify the nearest city for each latitude and longitude combination.
for coordinate in lats_lngs:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name

    # If the city is unique, then we will add it to the cities list.
    if city not in cities:
        cities.append(city)
        
# Print the city count 
len(cities)
```
